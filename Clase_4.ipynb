### Ejercicio 1

Extraer el título de la página de python.org 

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
 
html = urlopen("https://www.python.org/")

res = BeautifulSoup(html.read(),"html5lib")
 
print(res.title.text)
```

```python
import requests

r = requests.get("https://www.python.org/")

res = BeautifulSoup(r.text,"html5lib")
 
print(res.title.text)
```

### Ejercicio 2

Extraer departamentos del portal inmobiliario

```python
import requests

r = requests.get("https://www.portalinmobiliario.com/venta/departamento/villa-alemana-valparaiso?tp=2&op=1&ca=1&ts=1&dd=0&dh=6&bd=0&bh=6&or=&mn=1&sf=1&sp=0&ph=70.000.000")

res = BeautifulSoup(r.text,"html5lib")
 
print(res.title.text)
```

```python
deptos = res.find_all('div', {'class': 'col-sm-9 product-item-data'})
print(len(deptos))
for d in deptos:
    summary = d.find('div', {'class': 'product-item-summary'})
    print(summary.find('h4').text)
    for p in summary.find_all('p'): 
        print(p.text)
    for p in d.find_all('p', {'class': 'product-property'}): 
        print(p.text)
    print()
```

### Ejercicio 3

Extraer productos y precios de linio

```python
import requests
productos = []
for i in range(3):
    r = requests.get("https://www.linio.cl/c/computacion/pc-portatil?page=" + str(i+1))
    res = BeautifulSoup(r.text,"html5lib")
    products = res.find_all('div', {'class':'catalogue-product row'})
    for p in products:
        title = p.find('meta', {'itemprop':'name'})
        price = p.find('meta', {'itemprop':'price'})
        if title:
            print(title['content'])
        if price:
            print(price['content'])
        productos.append({'title':title['content'], 'price':price['content']})
```

```python
import pandas as pd
pd.DataFrame(productos)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://google.cl')

elem = driver.find_element_by_name('q')
elem.clear()
elem.send_keys("Magister en data science")
elem.send_keys(Keys.RETURN)

# WebDriverWait(driver, 3)
driver.implicitly_wait(10)

elems = driver.find_elements_by_css_selector('div.g')
for e in elems:
    print(e.find_element_by_tag_name('h3').text)
    print(e.find_element_by_tag_name('cite').text)
    print()
    
driver.close()
```
